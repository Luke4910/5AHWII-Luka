```python
import pandas as pd
import numpy as np
import seaborn as sns

import plotly 
import plotly.graph_objects as go
from statsmodels.formula.api import ols
```

```python
cars = pd.read_csv("mtcars.csv")
```

```python
print(cars.columns.tolist())
```

```python
sns.regplot(x="disp", y="mpg", data=cars)
```

```python
sns.pairplot(cars,vars=["mpg","disp","hp","drat","wt","qsec"],)
```

```python
import statsmodels.formula.api as smf
```

```python
mod=smf.ols('mpg ~ wt ',data=cars).fit()
```

```python
mod.summary()
```

```python
mod_multiple=smf.ols('mpg ~ wt+disp ',data=cars).fit()
```

```python
mod_multiple.summary()
```
